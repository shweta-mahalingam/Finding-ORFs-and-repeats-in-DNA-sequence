```python
#Question 2:
#a) What are the lengths of the sequences in the file?
#b) What is the longest sequence?
#c) what is the shortest sequence?
#d) Is there more than one longest or shortest sequence?
#e) What are their identifiers?

#subquestion a) length of sequences in the file (and shortest and longest sequence)
f = open("/content/dna.example.fasta","r")
file = f.readlines()

sequences = []
seq = ""
identifier = ""
iden = []

for f in file:
  if not f.startswith(">"):
    f = f.replace(" ","")
    f = f.replace("\n","")
    seq = seq+f
  else:
    f = f.replace ("marine metagenome JCVI_SCAF_1096627390048 genomic scaffold, whole genome shotgun sequence", "")
    f = f.replace("\n","")
    identifier = identifier+f
    iden.append(identifier)
    identifier = ""
    sequences.append(seq)
    seq = ""

sequences.append(seq)

sequences = sequences[1:]
print("The list of sequences", sequences)
print("The list of identifiers:", iden)
lengths = [len(i) for i in sequences]
print ("The length of sequences in the file are:", lengths)
print("The length of the longest sequence is:", max(lengths), "\n", "The length of the shortest sequence is:", min(lengths))

#list of sequence_length sorted
def sequences_sorted(list_sequences):
  list_of_sequences_sorted_length_wise = []
  sorted_sequences = sorted(list_sequences, key = len)
  for l in sorted_sequences:
    list_of_sequences_sorted_length_wise.append(len(l))
  return print ("This list is sorted length wise:", list_of_sequences_sorted_length_wise)

sequences_sorted(sequences)


#Subquestion b) Shortest and longest sequence and c) Is there more than one longest or shortest sequence?
dict_seq_and_length = {sequences[i]:lengths[i] for i in range(len(sequences))}
print("This is a dictionary with sequence and its length pair:", dict_seq_and_length)

val_min = min(lengths)
val_min_list = []

for key in dict_seq_and_length:
  if dict_seq_and_length[key] == val_min:
    lowest = key
    val_min_list.append(len(lowest))

print("The shortest sequence length is:", min(lengths), "\n", "This is list of shortest sequences:", val_min_list)

if len(val_min_list) >1:
  print("There are more than one shortest sequence")
  print("the list of shortest sequences are", +str(val_min_list))
else:
  print("The shortest DNA seq is " +str(lowest))

val_max = max(lengths)
val_max_list = []
for keyyy in dict_seq_and_length:
  if dict_seq_and_length[keyyy] == val_max:
    highest = keyyy
    val_max_list.append(len(highest))
print("The longest sequence length is:", max(lengths),"\n", "This is a list of longest sequences:", val_max_list)

if len(val_max_list) >1:
  print("There are more than one longest sequence")
  print("the list of longest sequences are", +str(val_max_list))
else:
  print("This is the longest sequence:", highest)


#Creating dictionary for identifier: sequence, length of sequence
dict_iden_seq_len = {iden[i]: [lengths[i], sequences[i]] for i in range(len(iden))}
print(dict_iden_seq_len)


#subquestion e) match input seq with idenifier for all shortest and longest sequence
def retrieve_iden(dict, n_list):
  for key, value in dict.items():
    for each_min in n_list:
      if each_min in value:
        print("This is the identifier:", key)


retrieve_iden(dict_iden_seq_len,  val_max_list)
retrieve_iden(dict_iden_seq_len, val_min_list)

```

```python
#Question 3: a) Given an input reading frame on the forward strand (1, 2, or 3) your program should be able to identify all ORFs present
# b) what is the length of the longest ORF in the file?
# c) What is the identifier of the sequence containing the longest ORF?
# d) For a given sequence identifier, what is the longest ORF contained in the sequence represented by that identifier?
# e) What is the starting position of the longest ORF in the sequence that contains it?

# Subquestion a) All the ORFs
def find_orf(sequences):
    # Find all ATG indexs
    start_position = 1
    start_indexs = []
    stop_indexs = []

    for i in range(0, len(sequences), 3):
        if sequences[i:i+3] == "ATG":   #find all start codons
            start_indexs.append(i)


    for i in range(0, len(sequences), 3):
        stops =["TAA", "TGA", "TAG"]
        if sequences[i:i+3] in stops:   # Find all stop codon indexs
            stop_indexs.append(i)

    orf = []
    mark = 0
    for i in range(0,len(start_indexs)):
        for j in range(0, len(stop_indexs)):
            if start_indexs[i] < stop_indexs[j] and start_indexs[i] > mark:
                orf.append(sequences[start_indexs[i]:stop_indexs[j]+3])
                mark = stop_indexs[j]+3


    return orf

orf_length = []
orf_read =[]
seq_orf_list = []
seq_orf_len = []
for i in sequences:
    orfs = find_orf(i)


    for j in orfs:

      orf_length.append(len(j))
      orf_read.append(j)

    df = {i:orfs}
        #print(df)
    seq_orf_list.append(df)

print("This is length of ORFS:", orf_length,  len(orf_length))
print("This is list of ORFs:", orf_read,  len (orf_read))
ORFread_len_dict = {orf_length[i]:orf_read[i] for i in range(len(orf_read))}
print("This is orf dict 1:", ORFread_len_dict, len(ORFread_len_dict))

y = max(orf_length)
max_orf_length = ORFread_len_dict.get(y)
print("This is longest ORF in the given file:", max(orf_length), ":", max_orf_length)
```
